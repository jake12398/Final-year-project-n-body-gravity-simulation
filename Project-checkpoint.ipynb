```python
def calculate_most_bound_stars(x, y, z, vx, vy, vz, m):
 
    N = len(x)
    most_bound_to = []
 
    G = 6.67e-11  # Gravitational constant
 
    for i in range(N):
        max_bound_energy = 1e300  # Initialize maximum bound energy to positive infinity
        most_bound_star_index = -1  # Initialize index of the most bound star to -1 (no star)
        for j in range(N):
            if i != j:
                r = np.sqrt((x[i] - x[j])**2 + (y[i] - y[j])**2 + (z[i] - z[j])**2)
                pe = -G * m[i] * m[j] / r
                ke = 0.5 * m[i] * (vx[i]**2 + vy[i]**2 + vz[i]**2)
                bound_energy = pe + ke
                # Update most bound star if bound energy is the most negative
                if bound_energy < max_bound_energy:
                    max_bound_energy = bound_energy
                    most_bound_star_index = j
        most_bound_to.append(most_bound_star_index)
 
    return most_bound_to
 

 
most_bound_to = calculate_most_bound_stars(x, y, z, vx, vy, vz, m)
 
# Print most bound star for each star
for i in range(N):
    star_index= i+1
    most_bound_star_index=most_bound_to[i]+1 if most_bound_to[i]!=-1 else "none"
    print(f"Star {star_index} is most bound to {most_bound_star_index}")
```

```python
def potential_energy(x, y, z, m, N):
    G = 6.67e-11
    Ep = np.zeros(N)
    for i in range(N):
        for j in range(N):
            if i != j:
                r = np.sqrt((x[i] - x[j])**2 + (y[i] - y[j])**2 + (z[i] - z[j])**2)
                Ep[i] += -0.5*(G * m[i] * m[j] / r )
    return Ep



def kinetic_energy(vx, vy, vz, m, N):
    K = np.zeros(N)
    for i in range(N):
        K[i] += 0.5 * m[i] * (vx[i]**2 + vy[i]**2 + vz[i]**2)
    return K
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

# ======================
# CONSTANTS
# ======================
G = 6.67e-11 * 1e3   # slight boost for visibility

N = 30
radius = 5e13
ep = 1e11

dt = 60 * 60 * 24 / 10  # timestep
max_steps = 20000

num_snapshots = 6

# ======================
# INITIAL CONDITIONS
# ======================
def generate_positions(n, r):
    pos = np.zeros((n, 3))
    i = 0
    while i < n:
        x = random.uniform(-r, r)
        y = random.uniform(-r, r)
        z = random.uniform(-r, r)
        if x*x + y*y + z*z <= r*r:
            pos[i] = (x, y, z)
            i += 1
    return pos


def generate_masses(n):
    return np.random.lognormal(mean=0.2, sigma=0.5, size=n) * 2e30


stars = generate_positions(N, radius)
x, y, z = stars[:, 0], stars[:, 1], stars[:, 2]

m = generate_masses(N)

# velocities (virialised system)
v_scale = np.sqrt(G * np.sum(m) / radius)

vx = np.random.normal(0, 1, N) * v_scale
vy = np.random.normal(0, 1, N) * v_scale
vz = np.random.normal(0, 1, N) * v_scale


# ======================
# GRAVITY FUNCTION
# ======================
def compute_acc(x, y, z, m):
    dx = x[:, None] - x[None, :]
    dy = y[:, None] - y[None, :]
    dz = z[:, None] - z[None, :]

    r2 = dx**2 + dy**2 + dz**2 + ep**2
    np.fill_diagonal(r2, np.inf)

    inv_r3 = 1.0 / (r2 * np.sqrt(r2))

    ax = -G * np.sum(m[None, :] * dx * inv_r3, axis=1)
    ay = -G * np.sum(m[None, :] * dy * inv_r3, axis=1)
    az = -G * np.sum(m[None, :] * dz * inv_r3, axis=1)

    return ax, ay, az


# ======================
# SNAPSHOT FUNCTION
# ======================
def save_snapshot(filename, x, y, z, vx, vy, vz, m, t):
    data = np.vstack((x, y, z, vx, vy, vz, m, t)).T
    np.savetxt(filename, data)


# ======================
# INITIAL PLOT 
# ======================
plt.figure(figsize=(7, 7))
plt.title("Initial State")

lim = np.max(np.abs(x)) * 1.2
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)

plt.scatter(x, y, s=20)
plt.show()


# ======================
# SNAPSHOT SETUP
# ======================
avg_v = np.mean(np.sqrt(vx**2 + vy**2 + vz**2))
endtime = (2 * radius) / (avg_v + 1e-10)

snapshot_interval = endtime / num_snapshots
last_snapshot_time = 0
snapshot_id = 1

time = 0


# save initial snapshot
save_snapshot(
    f"snapshot_{snapshot_id:04d}.data",
    x, y, z, vx, vy, vz, m,
    np.ones(N) * time
)
snapshot_id += 1


# ======================
# SIMULATION LOOP
# ======================
step = 0

while time < endtime and step < max_steps:

    ax, ay, az = compute_acc(x, y, z, m)

    # integration
    x += vx * dt + 0.5 * ax * dt**2
    y += vy * dt + 0.5 * ay * dt**2
    z += vz * dt + 0.5 * az * dt**2

    vx += ax * dt
    vy += ay * dt
    vz += az * dt

    time += dt
    step += 1

    # snapshot saving
    if time - last_snapshot_time >= snapshot_interval:
        save_snapshot(
            f"snapshot_{snapshot_id:04d}.data",
            x, y, z, vx, vy, vz, m,
            np.ones(N) * time
        )
        snapshot_id += 1
        last_snapshot_time = time


# ======================
# FINAL PLOT
# ======================
plt.figure(figsize=(7, 7))
plt.title("Final State")

lim = np.max(np.abs(x)) * 1.2
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)

plt.scatter(x, y, s=20, c='red')
plt.show()

```
